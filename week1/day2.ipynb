```python


np.random.seed(42)

data = {
    "ticker":['AAPL']*5+['GOOGL']*5+['MSFT']*5,
    "date":pd.date_range('2025-04-01',periods=5).tolist()*3,
    #这里的tolist是转换成普通的python列表，*3表示重复三次
    "close":np.random.randn(15)*2+150,
    #这里是生成15个，标准差为2，均值为150的随机数
    "volume":np.random.randint(1000,10000,15)

}
df = pd.DataFrame(data)

df
```

```python
df_multi = df.set_index(['ticker','date']).sort_index()
#set_index这里是将ticker和date提取出来作为索引，而close和volume仍然作为列名，（原本的数据默认的索引为0，1，

df_multi
```

```python
task_4 = df_multi.loc[('AAPL',slice('2025-04-01','2025-04-03')),:]
#元组第一个是ticker，第二个是使用slice确定切割的范围，：表示取所有
task_4
```

```python
task_5 = df_multi.xs('2025-04-02',level = 'date')['close']
#xs表示直接跨层级提取‘2025-04-02’的数据，
task_5
```

```python
task_6 = df_multi.groupby(level = 0)['close'].apply(lambda x: x/x.iloc[0]-1)
#iloc是进行数数，
#.apply 是对每组你想操作的数据执行你想进行的操作
task_6
```

```python
import pandas as pd
import os

class DataProcessor1:
    def __init__(self):
        self.data_path = "./data/"
    
    def load_data1(self, return_list=True):
        df_list = []
        
        for file in os.listdir(self.data_path):
            if file.endswith(".csv"):
                file_path = os.path.join(self.data_path, file)
                
                # 关键：跳过第二行垃圾数据
                df = pd.read_csv(file_path, skiprows=[1])
                
                ticker = file.replace(".csv", "")
                df["ticker"] = ticker
                
                # 处理日期
                df['date'] = pd.to_datetime(df['Date'])
                
                # 数字列自动转换（pandas 会自动处理）
                # 如果还有问题，用这个：
                # df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
                
                
                df_list.append(df)
                #pd.concat返回的是一个DataFrame，而不是list,这里的list是指数据结构中的list
        
        return df_list if return_list else pd.concat(df_list, ignore_index=True)
    #这里返回的df_list中，每个元素都是一个DataFrame
    def align_price(self, df_list, price_col='Close', fill_method='ffill'):
        tickers = [df['ticker'].iloc[0] for df in df_list]
        #从每个列表中取出ticker，并且只用取一个就够了，所以用,iloc[0]
        
        combined = pd.concat(
            [df.set_index('date')[price_col] for df in df_list],
            axis=1,
            keys=tickers
        )
        
        if fill_method == 'ffill':
            combined = combined.ffill()
        
        combined.index = pd.to_datetime(combined.index)
        combined = combined.sort_index()
        
        return combined


if __name__ == '__main__':
    processor = DataProcessor1()
    
    df_list = processor.load_data1(return_list=True)
    aligned_price = processor.align_price(df_list, price_col='Close')
    print(aligned_price.head())
    print(f"\n数据形状: {aligned_price.shape}")
```
