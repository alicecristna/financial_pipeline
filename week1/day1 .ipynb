```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "ticker":['APPLE']*5+['GOOGLE']*5+['MSFT']*5,
    "date":pd.date_range('2025-04-01',periods=5).tolist()*3,
    #这里的tolist是转换成普通的python列表，*3表示重复三次
    "close":np.random.randn(15)*2+150,
    #这里是生成15个，标准差为2，均值为150的随机数
    "volume":np.random.randint(1000,10000,15)

}
df = pd.DataFrame(data)
```

```python
df['close_zscore'] = df.groupby('ticker')['close'].transform(
    lambda x : (x-x.mean())/x.std()
)
print(df[['ticker','date','close','close_zscore']])
#这里写两个方括号是一次取多列的固定写法
```

```python
df_close_filter = df.groupby('ticker').filter(
    lambda x: x['volume'].mean()>5000
)
print(df_close_filter)
```

```python
df_aggs = df.groupby('ticker').agg({
    "close":['mean','max'],
    "volume":'sum'
})

print(df_aggs)
```
